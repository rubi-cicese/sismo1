```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as patches
import matplotlib.path as path
import matplotlib as mpl
```

```
data=pd.read_csv('data-02_IRIS.csv',names=['depth','mag','fecha','N'],skiprows=18,sep=',',engine='python')
print(data)
```

```
fig= plt.figure(figsize=(20,10))
plt.style.use('bmh')
n=plt.hist(data['N'],100,facecolor='g',edgecolor='black')
plt.ylabel=('Número de sismos ')
plt.xlabel('Tiempo transcurrido desde M7.2 (días)')
plt.title('Histograma de ocurrencia')
_=plt.show
```
